```python
template = """
Two persons are talking about legislation of euthanasia and death with dignity.
Base the conversation on what the two people were talking about and
provide three turns for each speaker to complete the dialogue.
Please tailor the tone and mannerisms of each speaker accordingly.
Answer the question in <{output_language}>.
YOU MUST USE <{output_language}> TO ANSWER THE QUESTION.
Question:"""
```

```python
system_message_prompt = SystemMessagePromptTemplate.from_template(template)
human_template = "{text}"
human_message_prompt = HumanMessagePromptTemplate.from_template(human_template)

chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt, human_message_prompt])

chat_prompt.format_messages(output_language="Korean",
                            text=text)
```

```python
query = chat_prompt.format_messages(input_language = "Korean",
                                    output_language = "Korean",
                                    text = text)
```

```python
chat_model.predict_messages(query)
```

```python
def get_completion(prompt, model="gpt-4", temperature=0.7, max_tokens=150, top_p=1.0, frequency_penalty=0, presence_penalty=0):
    try:
        response = openai.ChatCompletion.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            temperature=temperature,
            max_tokens=max_tokens,
            top_p=top_p,
            frequency_penalty=frequency_penalty,
            presence_penalty=presence_penalty
        )
        if response.choices:
            return response.choices[0].message["content"]
        else:
            return "No response from the model."
    except Exception as e:
        return f"Error: {str(e)}"
```
