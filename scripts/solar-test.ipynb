```python
from openai import OpenAI
import requests
```

```python
def load_api_key(file_path):
    with open(file_path, 'r') as file:
        api_key = file.read().strip()
    return api_key
```

# Mini-chat Model

```python
def get_solar_completion(api_key, prompt):
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    payload = {
        "model": "solar-1-mini-chat",
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]
    }

    try:
        response = requests.post("https://api.upstage.ai/v1/solar/chat/completions", json=payload, headers=headers)
        response_data = response.json()
        if response.status_code == 200:
            messages = response_data.get('choices', [{}])[0].get('message', {}).get('content', "No response from the model.")
            return messages
        else:
            return f"Error: {response.status_code}, {response_data.get('error', 'Unknown error')}"
    except Exception as e:
        return f"Error: {str(e)}"
```
